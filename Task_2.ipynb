```python
# 1. Working with Python data structures
# Lists, Dictionaries, Tuples, Sets

# List
grades = [85, 90, 78, 92, 88]

# Dictionary
student_scores = {"Alice": 85, "Bob": 90, "Charlie": 78, "David": 92, "Eva": 88}

# Tuple
students = ("Alice", "Bob", "Charlie", "David", "Eva")

# Set
unique_scores = set(grades)

print("List:", grades)
print("Dictionary:", student_scores)
print("Tuple:", students)
print("Set:", unique_scores)
```

```python
# 2. Functions for data processing
def average_score(scores):
    return sum(scores)/len(scores)

def top_student(scores_dict):
    return max(scores_dict, key=scores_dict.get)

print("Average grade:", average_score(grades))
print("Top student:", top_student(student_scores))
```

```python
# 3. Read and write CSV, JSON, and text files

import csv
import json

# Writing CSV
with open("grades.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Student", "Score"])
    for student, score in student_scores.items():
        writer.writerow([student, score])

# Reading CSV
with open("grades.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)  # Skip header
    for row in reader:
        print(row)

# Writing JSON
with open("grades.json", "w") as file:
    json.dump(student_scores, file)

# Reading JSON
with open("grades.json", "r") as file:
    data = json.load(file)
    print(data)

# Writing Text
with open("grades.txt", "w") as file:
    for student, score in student_scores.items():
        file.write(f"{student}: {score}\n")

# Reading Text
with open("grades.txt", "r") as file:
    print(file.read())
```

```python
# 4. List comprehension for data transformation
# Example: Increase all grades by 5%
updated_grades = [score*1.05 for score in grades]
print("Updated grades:", updated_grades)
```

```python
# 5. Error and exception handling
def safe_divide(a, b):
    try:
        return a/b
    except ZeroDivisionError:
        print("Error: Division by zero")
        return None

print("Safe divide 10/2:", safe_divide(10,2))
print("Safe divide 10/0:", safe_divide(10,0))
```
